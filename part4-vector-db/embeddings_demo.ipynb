```python
sentences = [
    # Cricket
    "The batsman scored a century in the match",
    "The bowler took a hat-trick in the final over",
    "The team won the match by chasing a huge target",

    # Cooking
    "Add spices and cook the vegetables for 10 minutes",
    "The recipe requires fresh ingredients and proper timing",
    "Boil the pasta and prepare the sauce separately",

    # Cybersecurity
    "Use strong passwords to protect your accounts",
    "Encryption helps secure sensitive data from hackers",
    "Firewalls prevent unauthorized access to networks",

    # Extra (to make total 10)
    "The chef prepared a delicious three-course meal"
]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

similarity_matrix = cosine_similarity(embeddings)

plt.figure(figsize=(8,6))
sns.heatmap(similarity_matrix, annot=True, fmt=".2f")
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 most similar sentences
top_indices = similarities.argsort()[-2:][::-1]

for idx in top_indices:
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {similarities[idx]:.4f}")
    print()
```
